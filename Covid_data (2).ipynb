```python
#Data set from the confirmed cases
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
data = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/refs/heads/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv")
data.head(15)
```

```python
deaths = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/refs/heads/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv")
recovered = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/refs/heads/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv")
```

```python
#How many times the contry reported in the dataset (false value)
data["Country/Region"].value_counts()
```

```python
# To have a true value we can sum the number of the confirmed covid's cases for each contry
confirmed_by_cases = data.groupby("Country/Region").sum(numeric_only=True)
confirmed_by_cases = confirmed_by_cases.drop(columns=["Lat","Long"])
#Slicing the first two info cause they are lat and long and plot
confirmed_by_cases.loc["China"].plot()
plt.title("Confirmed Cases in China")
plt.show()
```

```python
#top 10 countries with confirmed cases // grouping all the values of the top countries by .iloc[:,-1] and sorting them with .sort_values(ascending=False) non post-order
top_countries = confirmed_by_cases.iloc[:, -1].sort_values(ascending=False).head(10)

#plot config

plt.title("Top 10 Countries with Confirmed Cases")
plt.xlabel("Confirmed Cases")
plt.ylabel("Countries")
plt.barh(top_countries.index, top_countries)
plt.show()
```

```python
#Diferrence in the new cases number day by day
new_cases_india = confirmed_by_cases.loc["India"][2:].diff().dropna() #[2:] because lo and la
new_cases_india.plot()
```

```python
new_cases_us = confirmed_by_cases.loc["US"][2:].diff().dropna()
new_cases_us.plot()
```

```python
# Top countries cases
for c in top_countries.index:
    top_countries_cases = confirmed_by_cases.loc[c]
    top_countries_cases.plot()
plt.legend()
plt.title("Top 10 Countries with Confirmed Cases")
plt.xlabel("Confirmed Cases")
plt.ylabel("Countries")
plt.show()
```

```python
#Last date of confirmed cases
last_date = "3/9/23"
plt.title("Confirmed by cases 'Last date'")
confirmed_by_cases[last_date].sort_values(ascending=True).head(25).plot(kind="barh",color='green')
plt.figure(figsize=(10,5))
plt.show()
```

```python
deaths.head()
deaths.set_index("Province/State")
```

```python
recovered["Country/Region"].unique()
```

```python
def last_date(data):
  '''
   return sum of the data about the last date from all the data -> 3/9/23
  '''
  return data.groupby("Country/Region").sum().iloc[:,-1]

information_last_date = last_date(recovered),last_date(deaths),last_date(data)

information_covid_combined = pd.concat(information_last_date,axis=1)
information_covid_combined.columns = ["recovered","deaths","confirmed"]
information_covid_combined
```

```python
confirmed_deaths = deaths.groupby("Country/Region").sum(numeric_only=True)
top_d = confirmed_deaths.iloc[:,-1].sort_values(ascending=False).head(10)
fig = px.bar(deaths,x=top_d.index,y=top_d).show()
```

```python
sum_combined = information_covid_combined.sum()
lethal_rate = sum_combined["deaths"] / sum_combined["confirmed"] #lethal rate -> L = D/C * 100
print(lethal_rate * 100)
```

```python
lethal_country = information_covid_combined["deaths"] / information_covid_combined["confirmed"] * 100 #By country
information_covid_combined["lethal_country"] = lethal_country
information_covid_combined.head()
```

```python
lethal_rate_top_10 = information_covid_combined.sort_values("lethal_country", ascending=False).head(10)
plt.figure(figsize=(15,15))
plt.bar(lethal_rate_top_10.index,lethal_rate_top_10["lethal_country"])
```
